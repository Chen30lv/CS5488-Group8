```python
import numpy as np

import pandas as pd

import matplotlib as mpl

import matplotlib.pyplot as plt


import seaborn as sns
```

read data_info_main_par.csv
湿度我暂时没想到什么可以分析的，所以先就做了基础的展示
图表的背景和线条颜色可以根据ppt的风格再进行修改

```python
humidity_data = pd.read_csv('../2019-2023full_data/data_info_main_par.csv', 
                   header=0, names=["Date", "Relative Humidity"], usecols=[0,2])
```

2020 humidity data

```python
humidity_data_2020 = humidity_data[humidity_data["Date"].str.startswith("2020")]
humidity_data_2020["Date"] = pd.to_datetime(humidity_data_2020["Date"])
daily_data_2020 = humidity_data_2020.groupby(humidity_data_2020["Date"].dt.date)["Relative Humidity"].mean()


daily_humidity_data_2020 = pd.DataFrame({"Date": daily_data_2020.index, "Relative Humidity": daily_data_2020.values})
daily_humidity_data_2020['Date'] = daily_humidity_data_2020['Date'].astype(str).str[-5:]
#删除闰年
daily_humidity_data_2020.drop(59, inplace=True)
```

2021 humidity data

```python
humidity_data_2021 = humidity_data[humidity_data["Date"].str.startswith("2021")]
humidity_data_2021["Date"] = pd.to_datetime(humidity_data_2021["Date"])
daily_data_2021 = humidity_data_2021.groupby(humidity_data_2021["Date"].dt.date)["Relative Humidity"].mean()
daily_humidity_data_2021 = pd.DataFrame({"Date": daily_data_2021.index, "Relative Humidity": daily_data_2021.values})
daily_humidity_data_2021['Date'] = daily_humidity_data_2021['Date'].astype(str).str[-5:]
```

2022 humidity data

```python
humidity_data_2022 = humidity_data[humidity_data["Date"].str.startswith("2022")]
humidity_data_2022["Date"] = pd.to_datetime(humidity_data_2022["Date"])
daily_data_2022 = humidity_data_2022.groupby(humidity_data_2022["Date"].dt.date)["Relative Humidity"].mean()
daily_humidity_data_2022 = pd.DataFrame({"Date": daily_data_2022.index, "Relative Humidity": daily_data_2022.values})
daily_humidity_data_2022['Date'] = daily_humidity_data_2022['Date'].astype(str).str[-5:]
```

2023 humidity data

```python
humidity_data_2023 = humidity_data[humidity_data["Date"].str.startswith("2023")]
humidity_data_2023["Date"] = pd.to_datetime(humidity_data_2023["Date"])
daily_data_2023 = humidity_data_2023.groupby(humidity_data_2023["Date"].dt.date)["Relative Humidity"].mean()
daily_humidity_data_2023 = pd.DataFrame({"Date": daily_data_2023.index, "Relative Humidity": daily_data_2023.values})
daily_humidity_data_2023['Date'] = daily_humidity_data_2023['Date'].astype(str).str[-5:]
```

每日相对湿度折线图

```python

# 四个子图共享横坐标
fig, axs = plt.subplots(4, sharex=True, figsize = (25,20))

# 图表的背景颜色和折线颜色需要你根据ppt风格修改

# 在第一个子图中绘制2020年的每日湿度变化
x1 = daily_humidity_data_2020["Date"]
y1 = daily_humidity_data_2020["Relative Humidity"]

axs[0].plot(x1, y1, color = 'b')
axs[0].set_title('Hong Kong daily humidity in 2020', fontsize = 30)
axs[0].set_ylabel('Humidity', fontsize = 15)

# 在第二个子图中绘制2021年的每日湿度变化
x2 = daily_humidity_data_2021["Date"]
y2 = daily_humidity_data_2021["Relative Humidity"]

axs[1].plot(x2, y2, color = 'g')
axs[1].set_title('Hong Kong daily humidity in 2021', fontsize = 30)
axs[1].set_ylabel('Humidity', fontsize = 15)

# 在第三个子图中绘制2022年的每日湿度变化
x3 = daily_humidity_data_2022["Date"]
y3 = daily_humidity_data_2022["Relative Humidity"]

axs[2].plot(x3, y3, color = 'r')
axs[2].set_title('Hong Kong daily humidity in 2022', fontsize = 30)
axs[2].set_ylabel('Humidity', fontsize = 15)

# 在第四个子图中绘制2024年的每日湿度变化
x3 = daily_humidity_data_2023["Date"]
y3 = daily_humidity_data_2023["Relative Humidity"]

axs[3].plot(x3, y3, color = 'y')
axs[3].set_title('Hong Kong daily humidity in 2023', fontsize = 30)
axs[3].set_xlabel('Date', fontsize = 15)
axs[3].set_ylabel('Humidity', fontsize = 15)

# 显示图形
plt.xticks(range(0,len(x1),25),rotation=30, size=18)
plt.yticks(size=18)
plt.show()
```

```python
 
# 创建图形并分割为两个子图
fig_multiple, ax = plt.subplots(figsize = (25,14))

# 在第一个子图中绘制折线图选择需要的列
x1 = daily_humidity_data_2021['Date']
y1 = daily_humidity_data_2020["Relative Humidity"]
y2 = daily_humidity_data_2021["Relative Humidity"]
y3 = daily_humidity_data_2022["Relative Humidity"]
x4 = daily_humidity_data_2023['Date']
y4 = daily_humidity_data_2023["Relative Humidity"]

ax.plot(x1, y1, color = '#248BD6', label='2020')
ax.plot(x1, y2, color = '#83B8FF', label='2021')
ax.plot(x1, y3, color = '#C6CDFF', label='2022')
ax.plot(x4, y4, color = "#869BAD", label='2023')
ax.set_title('Hong Kong daily humidity in 2020 - 2023', fontsize = 30)
ax.set_xlabel('Date', fontsize = 20)
ax.set_ylabel('Humidity', fontsize = 20)
ax.set_facecolor("white")

# 显示图形
plt.xticks(range(0,len(x1),25),rotation=30, size=18)
plt.yticks(size=18)
plt.tight_layout()
# 添加图例
plt.legend()
plt.show()
```
